```python
import tkinter as tk
from tkinter import messagebox

def check_admission(jamb_score, num_credits, interview_status, department):

    interview_status = interview_status.lower()

    if interview_status not in ("pass", "fail"):
        print("Invalid interview status. Please enter 'pass' or 'fail'.")
        return False 
    




    if department == "Computer Science":
        if jamb_score >= 250 and num_credits >= 5 and interview_status == "pass":
            return True
        else:
            return False
    elif department == "Mass Communication":
        if jamb_score >= 230 and num_credits >= 5 and interview_status == "pass":
            return True
        else:
            return False
    else:
        print("Invalid department. Please enter 'Computer Science' or 'Mass Communication'.")
        return False




    
admitted_candidates = []
not_admitted_candidates = []

while True:
     department = input("Enter department (Computer Science or Mass Communication): ").strip().title()
     while department not in ("Computer Science", "Mass Communication"):
        print("Invalid department. Please enter 'Computer Science' or 'Mass Communication'.")
        department = input("Enter department (Computer Science or Mass Communication): ").strip().title()  
   
   
     jamb_score = int(input("Enter JAMB score: "))
     num_credits = int(input("Enter number of credits in 5 key subjects: "))
     interview_status = input("Enter interview status (pass/fail): ")


     is_admitted = check_admission(jamb_score, num_credits, interview_status, department)


     candidate_info = { "department": department,"jamb_score": jamb_score,"num_credits": num_credits,
     "interview_status": interview_status, }
     
     if is_admitted:
        admitted_candidates.append(candidate_info)
        print("Candidate is admitted to", department)
     else:
        not_admitted_candidates.append(candidate_info)
        print("Candidate is not admitted to", department)

        # Ask if the user wants to process another candidate
     while True:
        another_candidate = input("Do you want to process another candidate? (yes/no): ").lower()
        if another_candidate == "yes":
            break

        elif another_candidate == "no":
             print("Admitted Candidates:")
             for candidate in admitted_candidates:
                print(candidate)   
             print("\nNot Admitted Candidates:")
             for candidate in not_admitted_candidates:
                print(candidate)  # Print info 
             print("End of program.")
             exit() #Exit
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")
 
```
